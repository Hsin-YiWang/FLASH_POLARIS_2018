This log is part of the data_analysis for 1st POLARIS XFEL beamtime at Desy, FLASH (Oct.2018)

(1) We have some data saved in txt which is generated from our analyser computer. 

(2) The run number can be found from pur google sharging folder.

```python
import numpy as np
from matplotlib import pyplot as plt

runs = [37,38]

i = 0
while i < len(runs) :
    data=np.loadtxt("/Users/hsin-yiwang/Desktop/Insitucell_for_XPS/FlashBeamtime/DATA/nonHDF5/0%03d.txt"%runs[i])
    X=(data[:,0])
    Y=(data[:,1]) 
    plt.plot(X, Y, label = "run0%03d.txt" %runs[i])
    plt.legend(frameon=False,loc='upper right', prop={'size': 12})
    plt.xlabel('Kinetic energy (eV)',size=12)
    plt.ylabel('Intensity (a. u.)',size=12)
    #i = i+1 use this if you want to plot all figures together  
    plt.show()
    i = i+1
```
